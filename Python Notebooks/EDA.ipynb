```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar
```

```python
df = pd.read_csv('../Data/Cleaned_Data.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(['Recnum', 'Date'], inplace=True)
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['Month'] = df['Date'].dt.month
df['Weekend'] = (df['Day_of_Week'] >= 5).astype(int)
df['Year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.day
```

```python
frauds_df = df[df['Fraud']==1]
```

```python
df.sample(5)
```

```python
fraud = df['Fraud'].value_counts()
bars = plt.bar(fraud.index,fraud.values, width=0.8,align='center')
for bar, value in zip(bars, fraud.values):
    plt.text(bar.get_x() + bar.get_width()/2, 
             value, 
             f'{value}', 
             ha='center', 
             va='bottom')
plt.xlabel('Fraud')
plt.ylabel('Count')
plt.title('Data Imbalance')
plt.xticks([0, 1], ['Not Fraud', 'Fraud'])
plt.savefig('../Images/Data Imbalance.jpg')
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist(df['Merch state'], bins=len(df['Merch state'].unique()) + 1, rwidth=0.8)
plt.xticks(rotation=90) 
plt.xlabel('Merchant State')
plt.ylabel('Count')
plt.title('Transactions Per State')
plt.tight_layout() 
plt.savefig('../Images/Transactions per state.jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist(frauds_df['Merch state'], bins=len(frauds_df['Merch state'].unique()) + 1, rwidth=0.8)
plt.xticks(rotation=90) 
plt.xlabel('Merchant State')
plt.ylabel('Count')
plt.title('Frauds Per State')
plt.tight_layout() 
plt.savefig('../Images/Frauds Per State.jpg')
plt.show()
```

```python
day_names = [calendar.day_name[i] for i in range(7)]
day_of_week = df['Day_of_Week'].value_counts()
plt.bar(day_of_week.index,day_of_week.values)
plt.xlabel('Days')
plt.ylabel('Count')
plt.title('Transactions by Day')
plt.xticks(range(0, 7), day_names,rotation = 45)
plt.savefig('../Images/Transactions by Day.jpg')
plt.show()
```

```python
day_of_week = frauds_df['Day_of_Week'].value_counts()
plt.bar(day_of_week.index,day_of_week.values)
plt.xlabel('Days')
plt.ylabel('Count')
plt.title('Frauds by Day')
plt.xticks(range(0, 7), day_names,rotation = 45)
plt.savefig('../Images/Fraud By Day.jpg')
plt.show()
```

```python
month = df['Month'].value_counts()
month_names = [calendar.month_name[i] for i in range(1,13)]
plt.bar(month.index,month.values)
plt.xlabel('Days')
plt.ylabel('Count')
plt.title('Transaction per Month')
plt.xticks(range(1, 13), month_names,rotation = 90)
plt.savefig('../Images/Transactions per Month.jpg')
plt.show()
```
